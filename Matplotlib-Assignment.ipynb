# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations. It provides a wide range of tools for creating plots, charts, and graphs, making it an essential tool for data visualization in scientific, engineering, and data analysis fields.

Five types of plots that can be created using Pyplot include:

Line Plot: Displays data points connected by straight lines, often used for time series data or continuous data points.

Scatter Plot: Represents individual data points as dots on a two-dimensional plane, useful for visualizing relationships between two variables.

Bar Plot: Represents categorical data with rectangular bars, where the length of each bar corresponds to the value of the category.

Histogram: Shows the distribution of a continuous dataset by dividing it into intervals and displaying the frequency of data points falling into each interval.

Pie Chart: Represents data as slices of a circle, where each slice's size corresponds to the proportion of the data it represents. Pie charts are typically used to show percentages of a whole.

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Scatter Plot: Represents individual data points as dots on a two-dimensional plane, useful for visualizing relationships between two variables.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)  
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("This is my x vs y graph")
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

The subplot() function in Matplotlib is used to create multiple plots within a single figure. It is particularly useful when you want to display several related plots together for comparison or to visualize different aspects of the data simultaneously. The syntax for subplot() is subplot(nrows, ncols, index), where nrows and ncols represent the number of rows and columns in the grid of subplots, and index represents the specific location of the subplot in the grid, counting from left to right and top to bottom starting from 1.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Data for line 1
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

# Adjust layout for better appearance
plt.tight_layout()

# Show the plot
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


Bar Plot: Represents categorical data with rectangular bars, where the length of each bar corresponds to the value of the category.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a graphical representation that provides a visual summary of the distribution of a dataset. It displays the median, quartiles, and possible outliers of the data in a compact manner. Box plots are particularly useful when comparing the distribution of multiple datasets or identifying the presence of outliers.

Box plots are used for the following purposes:

Visualizing Data Distribution: Box plots provide a quick and concise way to understand the spread and central tendency of the data.

Comparing Distributions: Box plots allow easy comparison of the distribution of multiple datasets, making it useful for analyzing the differences between groups or categories.

Identifying Outliers: Box plots help in identifying potential outliers that are far from the typical range of the data.

Handling Skewed Data: Box plots are robust against skewed data and do not get affected by extreme values as much as other plots like histograms.

Summarizing Data: Box plots offer a clear and intuitive summary of the data's key statistical properties, making it easier to communicate and interpret the data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data generation
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a single list for plotting
data = [box1, box2]

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the box plot
ax.boxplot(data)

# Set labels for each box in the plot
ax.set_xticklabels(['Box 1', 'Box 2'])

# Set a title for the plot
plt.title('Box Plot Comparison')

# Show the plot
plt.show()
```
